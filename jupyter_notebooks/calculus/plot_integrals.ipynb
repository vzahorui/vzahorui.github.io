```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import seaborn as sns

%matplotlib inline
```

Example of finding area under a curve 

```python
with sns.axes_style("white"):
    fig, ax = plt.subplots(figsize=(12,6))


a, b = 10, 50  # integral limits
x = np.linspace(0, 60, 600)
def f(x):
    return 0.16*x**3-15*x**2+400*x

ax.plot(x, f(x), color='royalblue')
ax.set_xlim(0, 60)
ax.set_ylim(bottom=0)

# plot area
ix = np.linspace(a, b)
iy = f(ix)
verts = [(a, 0), *zip(ix, iy), (b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5', zorder=1)
ax.add_patch(poly)
ax.plot([a, a], [3400, 0], color='grey', linestyle='--')
ax.plot([b, b], [3400, 0], color='grey', linestyle='--')

#plot bars
bar_width = (b-a)/10
idx = np.array([10, 14, 18, 22, 26, 30, 34, 38, 42, 46])
ax.bar(idx+bar_width/2, f(idx), 
       width=bar_width, color='indianred', edgecolor='gray', alpha=0.3, zorder=5)

d_x1 = a + bar_width/2
d_x2 = a + bar_width/2 + bar_width
d_xn = b - bar_width/2

ax.set_xticks((d_x1, d_x2, d_xn))
ax.set_xticklabels(('$\Delta x_1$', '$\Delta x_2$', '$\Delta x_n$'), fontsize=13)
ax.annotate('$a$', xy=(0.16, -0.05), ha='center', va='center', xycoords='axes fraction', fontsize=15)
ax.annotate('$b$', xy=(0.84, -0.05), ha='center', va='center', xycoords='axes fraction', fontsize=15)
ax.set_yticks([])

ax.set_ylabel('$f(x)$', fontsize=15, rotation=0, labelpad=20)
plt.savefig('../../assets/images/calculus/plot_area_under_curve.png', bbox_inches='tight');
```

fundamental theorem of calculus 

```python
with sns.axes_style("white"):
    fig, ax = plt.subplots(figsize=(12,6))


a, x_, b = 25, 49, 72  # limits
x = np.linspace(8, 90, 600)
def f(x):
    return (x-6)**2*5+(x-9)**2*6-(x-50)**3+40000
ax.plot(x, f(x), color='royalblue')
ax.set_xlim(8, 90)
ax.set_ylim(0, 130000)

# plot area
ix = np.linspace(a, b)
iy = f(ix)
verts = [(a, 0), *zip(ix, iy), (b, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5', zorder=1)
ax.add_patch(poly)
ax.plot([a, a], [f(a), 0], color='grey', linestyle='--')
ax.plot([b, b], [f(b), 0], color='grey', linestyle='--')
ax.plot([x_, x_], [f(x_), 0], color='darkslategrey', linestyle='--')

ax.set_xticks((a, x_, b))
ax.set_xticklabels(('$a$', '$x$', '$b$'), fontsize=14)
ax.set_yticks([])

ax.annotate('$t$', xy=(0.98, -0.05), ha='center', va='center', xycoords='axes fraction', fontsize=15)
ax.annotate('$y$', xy=(-0.02, 0.93), ha='center', va='center', xycoords='axes fraction', fontsize=15)

ax.annotate('$y = f(x)$', xy=(0.73, 0.62), ha='center', va='center', xycoords='axes fraction', fontsize=17)
plt.savefig('../../assets/images/calculus/plot_area_under_curve_with_middle_value.png', bbox_inches='tight');
```

improper integral

```python
with sns.axes_style("white"):
    fig, ax = plt.subplots(figsize=(12,6))


a, x_, b = 25, 49, 72  # limits
x = np.linspace(0.00001, 4, 600)
def f(x):
    return 1/x**2
ax.plot(x, f(x), color='royalblue')
ax.set_xlim(0.3, 4)
ax.set_ylim(0, 6)

# plot area
ix = np.linspace(1, 4)
iy = f(ix)
verts = [(1, 0), *zip(ix, iy), (4, 0)]
poly = Polygon(verts, facecolor='0.9', edgecolor='0.5', zorder=1)
ax.add_patch(poly)
ax.plot([1, 1], [1, 0], color='grey', linestyle='--')

ax.set_xticks([1])
ax.xaxis.set_tick_params(labelsize=12)
ax.set_yticks([])

ax.set_ylabel('$f(x)$', fontsize=15, rotation=0, labelpad=20)
plt.savefig('../../assets/images/calculus/plot_area_under_improper_integral.png', bbox_inches='tight');
```
