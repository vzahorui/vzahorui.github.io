```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(figsize=(12,6))
    
x = np.linspace(0, 5, 600)
def f(x):
    return 10/(np.exp(x*(-2))*((10-1)/1)+1)

ax.plot(x, f(x), color='royalblue')
ax.plot([0, 5], [10, 10], color='grey', linestyle='--', dashes=(5,5))

ax.set_xlim(0, 5)
ax.set_ylim(bottom=0)

ax.tick_params(axis='both', which='major', labelsize=12)

ax.annotate('$t$', xy=(0.98, -0.08), ha='center', va='center', xycoords='axes fraction', fontsize=15)
ax.annotate('$f(x)$', xy=(-0.07, 0.5), ha='center', va='center', xycoords='axes fraction', fontsize=15)

plt.savefig('../../assets/images/calculus/plot_logistic_equation.png', bbox_inches='tight');
```
