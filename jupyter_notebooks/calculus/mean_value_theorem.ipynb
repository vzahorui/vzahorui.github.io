```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
X = np.linspace(0, 52, 3000)

def func(x):
    return 0.16*x**3-18*x**2+480*x

with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(figsize=(12,7))
    
ax.set_xlim(0, 52)
ax.set_ylim(-2000, 5200)

plt.plot(X, func(X), color='royalblue') # the actual function
plt.scatter([7, 40], [func(7), func(40)], color='firebrick', zorder=20)
plt.text(5.3, func(7)+300, r'$f(a)$', fontsize=15)
plt.plot([7, 7], [-2000, func(7)], linestyle='--', color='grey', dashes=(5,5))
ax.annotate('$a$', xy=(0.14, -0.05), ha='center', va='center', xycoords='axes fraction', fontsize=15)
plt.text(40, func(40)+200, r'$f(x)$', fontsize=15)
plt.plot([40, 40], [-2000, func(40)], linestyle='--', color='grey', dashes=(5,5))
ax.annotate('$x$', xy=(0.79, -0.05), ha='center', va='center', xycoords='axes fraction', fontsize=15)

slope = (func(40)-func(7))/33
intercept1 = func(7) - 7*slope
plt.plot(X, X*slope+intercept1, color='peru') # secant line
plt.plot(X, X*slope+intercept1+1900, color='dimgrey', linestyle='--')
plt.scatter([21], [func(21)], color='firebrick', zorder=20)
plt.text(22, func(21)+200, r'$f(c)$', fontsize=15)
plt.plot([21, 21], [-2000, 21*slope+intercept1+1900], linestyle='--', color='grey', dashes=(5,5))
ax.annotate('$c$', xy=(0.42, -0.05), ha='center', va='center', xycoords='axes fraction', fontsize=15)
plt.savefig('../../assets/images/calculus/mean_value_theorem.png', bbox_inches='tight');
```
