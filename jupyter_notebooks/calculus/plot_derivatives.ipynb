```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Limits

Some example function

```python
with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(figsize=(12,6))
x = np.linspace(-15, 15, 600)
y = 1/abs(x)
ax.plot(x,y)
ax.set_ylim(0, 6)
ax.set_xlim(-6, 6)
ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$f(x)$', fontsize=14, rotation=0, labelpad=20)
plt.savefig('../../assets/images/calculus/plot_1_div_x.png', bbox_inches='tight');
```

Types of discontinuity

```python
with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(1, 3, figsize=(20,6))
    
x0 = np.linspace(-15, 15, 600)
y0 = x0**2
ax[0].plot(x0, y0, color='royalblue', zorder=3)
ax[0].scatter(10, 100, facecolor='white', edgecolor='royalblue', s=70, zorder=8)
ax[0].scatter(10, 50, color='royalblue')
ax[0].set_ylim(-9, 210)

x11 = np.linspace(-15, 5, 600)
y11 = x11**2
x12 = np.linspace(5, 16.5, 600)
y12 = x12*(-11) + 200
ax[1].plot(x11, y11, color='royalblue', zorder=3)
ax[1].plot(x12, y12, color='royalblue', zorder=3)
ax[1].scatter(x11[-1], y11[-1], facecolor='white', edgecolor='royalblue', s=70, zorder=8)
ax[1].scatter(x12[0], y12[0], color='royalblue')
ax[1].set_ylim(-9, 210)
ax[1].set_xlim(-16.5, 16.6)

x21 = np.linspace(-17, 6.9999, 600)
x22 = np.linspace(7.00000001, 17, 600)
y21 = 10/(x21 - 7) + 5
y22 = 10/(x22 - 7) + 5
ax[2].plot(x21, y21, x22, y22,  color='royalblue')
ax[2].plot([7, 7], [-1000, 1000], '--', color='grey')
ax[2].set_ylim(-40, 40)
ax[2].set_xlim(-15, 17)
    
ax[0].set_title('Point/Removable discontinuity', fontsize=14)
ax[1].set_title('Jump discontinuity', fontsize=14)
ax[2].set_title('Asymptotic discontinuity', fontsize=14)
plt.savefig('../../assets/images/calculus/plot_discontinuity_types.png', bbox_inches='tight');
```

# Derivatives

Different slopes

```python
with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(figsize=(8,6))
    
x = np.linspace(-2, 16, 600)
def func(x):
    return 4*x + 2*x**2 + 3
ax.plot(x, func(x), color='royalblue')
ax.set_xlim(0, 12)
ax.set_ylim(-10, 320)

# plot points
ax.scatter(2, func(2), color='royalblue')
ax.annotate('A', xy=(2, func(2)+20), fontsize=18)

ax.scatter(4, func(4), color='royalblue')
ax.annotate('B', xy=(4, func(4)+20), fontsize=18)

ax.scatter(8, func(8), color='royalblue')
ax.annotate('C', xy=(8, func(8)+20), fontsize=18)

ax.scatter(10, func(10), color='royalblue')
ax.annotate('D', xy=(10, func(10)+20), fontsize=18)

# slopes
ax.plot([2, 4], [func(2), func(4)], color='peru', linewidth=4)
ax.plot([8, 10], [func(8), func(10)], color='peru', linewidth=4)
plt.savefig('../../assets/images/calculus/plot_slope.png', bbox_inches='tight');
```

Continuous function with not differential point

```python
with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(figsize=(10,6))
    
x1 = np.linspace(-9, -0.0000001, 600)
x2 = np.linspace(0.0000001, 9, 600)
y1 = x1 * (-2)
y2 = x2 * 2
ax.plot(x1, y1, x2, y2, color='royalblue', zorder=2)
ax.scatter(0, 0, color='peru', zorder=5, s=100)
ax.set_ylim(-1, 17)
ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$f(x)$', fontsize=14, rotation=0, labelpad=20)
plt.savefig('../../assets/images/calculus/plot_no_slope.png', bbox_inches='tight');
```

# Analyzing functions

Critical points

```python
with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(figsize=(10,6))

x1 = np.linspace(0, 70, 600)
y1 = 0.16*x1**3-18*x1**2+480*x1
x2 = np.linspace(70, 120, 600)
y2 = -400/(69-x2)-(x2-90)**3*0.05-520
ax.plot(x1, y1, x2, y2, color='royalblue')

# global maximum
ax.scatter(x1[np.argmax(y1)], y1.max(), color='royalblue')
ax.annotate('A', xy=(x1[np.argmax(y1)], y1.max() + 200), fontsize=18)
ax.plot([x1[np.argmax(y1)]-10, x1[np.argmax(y1)]+10], [y1.max(), y1.max()], linestyle = '--', color='peru')
ax.plot([x1[np.argmax(y1)], x1[np.argmax(y1)]], [-9000, y1.max()], linestyle = '--', dashes=(5,5), color='grey')

# local minimum
ax.scatter(x1[np.argmin(y1)], y1.min(), color='royalblue')
ax.annotate('B', xy=(x1[np.argmin(y1)]-1, y1.min() + 200), fontsize=18)
ax.plot([x1[np.argmin(y1)]-10, x1[np.argmin(y1)]+10], [y1.min(), y1.min()], linestyle = '--', color='peru')
ax.plot([x1[np.argmin(y1)], x1[np.argmin(y1)]], [-9000, y1.min()], linestyle = '--', dashes=(5,5), color='grey')

# local maximum (undefined derivative)
ax.scatter(x1[-1], y1[-1], color='royalblue')
ax.annotate('C', xy=(x1[-1], y1[-1]+200), fontsize=18)
ax.plot([x1[-1], x1[-1]], [-9000, y1[-1]], linestyle = '--', dashes=(5,5), color='grey')

# Another critical point - neither minimum nor maximum
ax.scatter(90, -400/(69-90)-(90-90)**3*0.05-520, color='royalblue')
ax.annotate('D', xy=(90, -400/(69-90)-(90-90)**3*0.05-520 + 200), fontsize=18)
ax.plot([90-12, 90+12], [-400/(69-90)-(90-90)**3*0.05-520, -400/(69-92)-(92-90)**3*0.05-520], linestyle = '--', color='peru')
ax.plot([90, 90], [-9000, -400/(69-90)-(90-90)**3*0.05-520], linestyle = '--', dashes=(5,5), color='grey')

# Absolue minima over the interval
ax.scatter(x2[-1], y2[-1], color='royalblue')
ax.annotate('E', xy=(x2[-1] - 5, y2[-1] - 200), fontsize=18)

ax.set_xlim(0, 120)
ax.set_ylim(-2200, 4400)
ax.set_xlabel('$x$', fontsize=14)
ax.set_ylabel('$f(x)$', fontsize=14, rotation=0, labelpad=20)   
plt.savefig('../../assets/images/calculus/plot_critical_values.png', bbox_inches='tight');
```

Concavity

```python
with sns.axes_style("whitegrid"):
    fig, ax = plt.subplots(3, 1, figsize=(10,12), sharex=True)

x = np.linspace(-3, 3.5, 600)
y = -x**3+x**2+7*x-6
y1 = -3*x**2+2*x+7
y2 = -6*x+2

ax[0].plot(x, y, color='royalblue')
ax[0].scatter([1.896805, -1.23], [4.0510479, -11.23623309], color='royalblue')
ax[0].annotate('A', xy=(-1.23, -11.23623309+2), fontsize = 18)
ax[0].annotate('B', xy=(1.896805, 4.0510479+2), fontsize = 18)

ax[1].plot(x, y1, color='royalblue')
ax[1].scatter([1.896805, -1.23, 0.33333], [0, 0, 7.3333], color='royalblue')
ax[1].annotate('A', xy=(-1.23, 0+2), fontsize = 18)
ax[1].annotate('B', xy=(1.896805, 0+2), fontsize = 18)
ax[1].annotate('C', xy=(0.33333, 7.3333+2), fontsize = 18)
ax[1].set_ylim(-27, 14)

ax[2].plot(x, y2, color='royalblue')
ax[2].scatter(0.33333, 0, color='royalblue')
ax[2].annotate('C', xy=(0.33333, 0+2), fontsize = 18)

ax[0].set_ylabel('$f(x)$', fontsize=14, rotation=0, labelpad=20)
ax[1].set_ylabel('$f\'(x)$', fontsize=14, rotation=0, labelpad=20)
ax[2].set_ylabel('$f\'\'(x)$', fontsize=14, rotation=0, labelpad=20)
ax[2].set_xlabel('$x$', fontsize=14)

for ax in ax:
    ax.set_xlim(-3, 3.5)
    ax.plot([-10, 10], [0, 0], color='grey', linewidth=1.2)
plt.savefig('../../assets/images/calculus/plot_concavity.png', bbox_inches='tight');
```
