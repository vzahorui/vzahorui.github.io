```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

# Vector addition and multiplication

```python
with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(1, 2, figsize=(14, 6))
    

ax[0].set_xlim(-5, 6)
ax[0].set_ylim(-1, 6)  
ax[0].quiver([0, -2, 0], # x origin
             [0, 4, 0], # y origin
             [-2, 6, 4], # x end
             [4, 1, 5], # y end
             color=['peru', 'olive', 'navy'], lw=0.7,
             angles='xy', scale_units='xy', scale=1)
ax[0].text(-2.1, 2, '$\overrightarrow{a}$', fontsize=17)
ax[0].text(0.9, 4.8, '$\overrightarrow{b}$', fontsize=17)
ax[0].text(2.8, 2.5, '$\overrightarrow{a}+\overrightarrow{b}$', fontsize=17)
ax[0].set_title('Vector addition', fontsize=15)


ax[1].set_xlim(-3, 14)
ax[1].set_ylim(-1, 4)
ax[1].quiver([0, 0], # x origin
             [0, 0], # y origin
             [12, 4], # x end
             [3, 1], # y end
             color=['peru', 'navy'], lw=0.7,
             angles='xy', scale_units='xy', scale=1)
ax[1].text(1.8, 0.8, '$\overrightarrow{a}$', fontsize=17)
ax[1].text(6.8, 1.4, r'$\overrightarrow{a} \times 3$', fontsize=17)
ax[1].set_title('Vector scaling', fontsize=15)

plt.tight_layout()
plt.savefig('../../assets/images/linear_algebra/simple_vectors.png', bbox_inches='tight');
```

# 3D vectors

```python
fig = plt.figure(figsize=(14,10))
ax = plt.axes(projection='3d')

start_point = np.array([0, 0, 0])

v1 = np.array([3, 5, 10])
v2 = np.array([4, 7, 20])
v3 = np.array([10, 6, 2])

ax.plot3D(*zip(start_point, v1), c='royalblue')
ax.plot3D(*zip(start_point, v2), c='olive')
ax.plot3D(*zip(start_point, v3), c='indianred')
ax.text(5, 2.5, 21, s='$\overrightarrow{A}$', fontsize=15)
ax.text(4, 2.5, 9, s='$\overrightarrow{B}$', fontsize=15)
ax.text(9, 2.5, 6, s='$\overrightarrow{C}$', fontsize=15)

plt.savefig('../../assets/images/linear_algebra/3d_vectors.png', bbox_inches='tight');
```

# Vector projection

```python
with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(figsize=(10, 6))
    
ax.set_xlim(0, 13)
ax.set_ylim(0, 6)

ax.annotate(s='', xy=(12, 0), xytext=(0, 0), arrowprops=dict(width=2, color='darkgrey'))
ax.annotate(s='', xy=(8, 0), xytext=(0, 0), arrowprops=dict(width=2, color='cornflowerblue'))
ax.annotate(s='', xy=(8, 5), xytext=(0, 0), arrowprops=dict(width=2, color='navy'))
ax.annotate(s='', xy=(8, 5), xytext=(8, 0), arrowprops=dict(width=2, color='indianred'))

ax.annotate(s='$\overrightarrow{a}$', xy=(0, 0), xytext=(4, 3), fontsize=18)
ax.annotate(s='$\overrightarrow{a_1}$', xy=(0, 0), xytext=(4, 0.3), fontsize=18)
ax.annotate(s='$\overrightarrow{a_2}$', xy=(0, 0), xytext=(8.2, 2.5), fontsize=18)
ax.annotate(s='$\overrightarrow{b}$', xy=(0, 0), xytext=(6, -0.8), fontsize=18)
ax.annotate(s=r'$\alpha$', xy=(0, 0), xytext=(1, 0.25), fontsize=16)

plt.savefig('../../assets/images/linear_algebra/vectors_and_cosine.png', bbox_inches='tight');
```
