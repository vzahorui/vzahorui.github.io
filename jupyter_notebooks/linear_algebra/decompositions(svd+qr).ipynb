```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(figsize=(10, 6))
    
ax.set_xlim(0, 10)
ax.set_ylim(0, 6)

ax.annotate(s='', xy=(0, 5), xytext=(0, 0), arrowprops=dict(width=2, color='indianred'))
ax.annotate(s='', xy=(8, 0), xytext=(0, 0), arrowprops=dict(width=2, color='blue'))
ax.annotate(s='', xy=(8, 5), xytext=(0, 0), arrowprops=dict(width=2, color='dimgrey'))

ax.plot([0, 8], [5, 5], color='grey', linestyle='--', dashes=(10, 8))
ax.plot([8, 8], [0, 5], color='grey', linestyle='--', dashes=(10, 8))

ax.annotate(s='$\overrightarrow{a}$', xy=(0, 0), xytext=(4, 3), fontsize=18)
ax.annotate(s='$\overrightarrow{v_2}$', xy=(0, 0), xytext=(-0.7, 2.5), fontsize=18)
ax.annotate(s='$\overrightarrow{v_1}$', xy=(0, 0), xytext=(4, -0.7), fontsize=18)
ax.annotate(s='$s_2$', xy=(0, 0), xytext=(8.2, 2.5), fontsize=18)
ax.annotate(s='$s_1$', xy=(0, 0), xytext=(4, 5.2), fontsize=18)

plt.savefig('../../assets/images/linear_algebra/simple_vector_decomposition.png', bbox_inches='tight');
```

```python
with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(figsize=(10, 10))

origin = ([0], [0])
V = np.array([[8, 6], [8,-6]]) # V[0] - original vector, V[1] - its reflection through hyperplane

v_magn = np.linalg.norm(V[0]) # magnitute of vector V[0]
v_norm = V[0]/np.linalg.norm(V[0]) # scaled version of V[0]

ax.quiver(*origin, V[:,0], V[:,1], angles='xy', scale_units='xy', scale=1, color=['royalblue', 'indianred']) # original vctor and its reflection through hyperplane
ax.axhline(0, linestyle='--') # plot hyperplane
ax.quiver(8, 6, 0, -12, angles='xy', scale_units='xy', scale=1, color='grey') # v vector - difference between reflection and original vector

ax.text(8.5, 1, '$\overrightarrow v$', fontsize=25, ha='center', va='center')
ax.text(3.6, 4, '$\overrightarrow a$', fontsize=25, ha='center', va='center')
ax.text(3.6, -4, '$\overrightarrow r$',  fontsize=25, ha='center', va='center')
ax.text(11, 0.4, '$mirror$',  fontsize=16, ha='center', va='center')

plt.xlim(-1, 12)
plt.ylim(-8, 8)

ax.set_yticklabels([])
ax.set_xticklabels([])

plt.savefig('../../assets/images/linear_algebra/vector_reflection.png', bbox_inches='tight');
```
