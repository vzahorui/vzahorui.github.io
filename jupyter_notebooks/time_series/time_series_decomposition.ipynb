```python
import numpy as np
import matplotlib.pyplot as plt

def create_time_series_components(num_points=365):
    """
    Creates synthetic time series components for demonstration.

    Args:
        num_points (int): The number of data points to generate.

    Returns:
        tuple: A tuple containing the time array, trend, seasonality,
               and random noise components.
    """
    # Create the time array (e.g., days)
    time = np.arange(num_points)

    # 1. Create a linear trend
    # A gradual upward slope
    trend = 0.5 * time + 50

    # 2. Create a seasonal pattern (e.g., yearly cycle)
    # Using a sine wave to simulate a repeating pattern
    seasonality = 25 * np.sin(2 * np.pi * time / 365) + 10 * np.sin(2 * np.pi * time / 90)

    # 3. Create random noise
    # Random fluctuations around the trend and seasonality
    rng = np.random.default_rng(42) # For reproducibility
    noise = rng.normal(loc=0, scale=15, size=num_points)

    return time, trend, seasonality, noise

def plot_time_series(time, trend, seasonality, noise):
    """
    Plots the time series and its components in a 2x2 grid.
    """
    # Create the final time series by adding the components
    series = trend + seasonality + noise

    # Create a figure with 2x2 subplots to show the breakdown
    fig, axes = plt.subplots(2, 2, figsize=(12, 8), sharex=True)
    fig.suptitle('Time Series Decomposition', fontsize=18)

    # Flatten the axes array for easy iteration
    axes = axes.flatten()

    # Plot the full time series
    axes[0].plot(time, series, color='blue')
    axes[0].set_title('Full Time Series', fontsize=14)
    axes[0].set_ylabel('Value')
    axes[0].grid(True, which='both', linestyle='--')

    # Plot the trend component
    axes[1].plot(time, trend, color='red')
    axes[1].set_title('Trend Component', fontsize=14)
    axes[1].set_ylabel('Value')
    axes[1].grid(True, which='both', linestyle='--')

    # Plot the seasonality component
    axes[2].plot(time, seasonality, color='green')
    axes[2].set_title('Seasonality Component', fontsize=14)
    axes[2].set_ylabel('Value')
    axes[2].grid(True, which='both', linestyle='--')

    # Plot the random noise component
    axes[3].plot(time, noise, color='gray')
    axes[3].set_title('Irregularity (Noise)', fontsize=14)
    axes[3].set_ylabel('Value')
    axes[3].grid(True, which='both', linestyle='--')

    for ax in axes:
        ax.set_xlabel("Time")

    plt.tight_layout(rect=(0, 0, 1, 0.96))
    plt.savefig('../../assets/images/time_series/times_series_components.png', bbox_inches='tight')
    plt.show()
 

time, trend, seasonality, noise = create_time_series_components()
plot_time_series(time, trend, seasonality, noise)
```
