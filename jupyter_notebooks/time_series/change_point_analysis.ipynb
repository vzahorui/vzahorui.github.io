```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
rng = np.random.default_rng(12345) # random number generator
```

```python
n1, n2, n3 = 10, 13, 8
size = 64
x11 = np.linspace(0, n1, size)
x12 = np.linspace(n1, n1+n2, size)
x13 = np.linspace(n1+n2, n1+n2+n3, size)

y11 = rng.normal(2, 0.3, size)
y12 = rng.normal(5, 0.3, size)
y13 = rng.normal(3, 0.3, size)

x1, y1 = np.concatenate([x11, x12, x13]), np.concatenate([y11, y12, y13])


x21 = np.linspace(0, n1, size)
x22 = np.linspace(n1, n1+n2, size)
x23 = np.linspace(n1+n2, n1+n2+n3, size)

y21 = rng.normal(3, 0.1, size)
y22 = rng.normal(3, 0.5, size)
y23 = rng.normal(3, 0.3, size)

x2, y2 = np.concatenate([x21, x22, x23]), np.concatenate([y21, y22, y23])
```

```python
with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(1, 2, figsize=(16, 7))
    
ax[0].plot(x1, y1, color='steelblue')
ax[0].vlines([n1, n1+n2], 0, 6, color='lightcoral')
ax[0].set_title('Change in mean', fontsize=15)

ax[1].plot(x2, y2, color='steelblue')
ax[1].vlines([n1, n1+n2], 0, 6, color='lightcoral')
ax[1].set_title('Change in variance', fontsize=15)
plt.tight_layout()
plt.savefig('../../assets/images/time_series/basic_change_point_examples.png', bbox_inches='tight');
```
