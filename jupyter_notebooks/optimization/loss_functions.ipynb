```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.random import default_rng

rng = default_rng()
%matplotlib inline
```

# Regression loss functions

```python
def mse(x):
    return np.square(x)

def mae(x):
    return np.absolute(x)

x = rng.standard_normal(95)*3
x.sort()

with sns.axes_style('darkgrid'):
    fig, ax = plt.subplots(figsize=(12, 7))
ax.scatter(x, mse(x), alpha=0.7, s=45, label='Squared error')
ax.scatter(x, mae(x), alpha=0.7, s=45, label='Absolute error')
plt.legend(loc='best')
ax.set_title('Squared error and absolute error functions', fontsize=15)
ax.set_ylim(-1, 8)
plt.ylabel('Total loss', fontsize=13)
plt.xlabel('Error', fontsize=13)
plt.savefig('../../assets/images/optimization/absolute_and_squared_error.png', bbox_inches='tight');
```

```python
x = rng.standard_normal(95)*5
x.sort()

def huber_m(x):
    sigma = np.std(x)
    k = 1.345
    y = np.where(np.abs(x)<=k, x**2/2, k*np.abs(x)-k**2/2)
    return y

def bisquare_m(x):
    sigma = np.std(x)
    k = 4.685
    y = np.where(np.abs(x)<=k, k**2/6*(1-(1-(x/k)**2)**3), k**2/6)
    return y

with sns.axes_style('darkgrid'):
    fig, ax = plt.subplots(figsize=(12, 7))
ax.scatter(x, huber_m(x), alpha=0.7, s=45, label='Huber M-estimator')
ax.scatter(x, bisquare_m(x), alpha=0.7, s=45, label='Bisquare M-estimator')
ax.scatter(x, mse(x), alpha=0.7, s=45, label='Squared error')
plt.legend(loc='best')
ax.set_title('M-estimator functions', fontsize=15)
ax.set_ylim(-1, 8)
plt.ylabel('Total loss', fontsize=13)
plt.xlabel('Error', fontsize=13)
plt.savefig('../../assets/images/optimization/m_estimators_loss.png', bbox_inches='tight');
```
