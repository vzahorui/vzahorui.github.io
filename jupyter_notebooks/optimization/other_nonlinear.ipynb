```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Nedler-Mead algorithm

```python
X = np.linspace(1, 50, 3000)

def func(x, a, b):
    return round(sum((np.cos(x*a)-b)**2)/10000, 2)
```

```python
with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(2, 2, figsize=(18,12))

for i in ax:
    for j in i:
        j.set_xlim(-30, 28)
        j.set_ylim(-48, 28)

A = (10, 20) # (-3, -8)
B = (-18, 1)
C = (12, -17)
ax[0][0].scatter([A[0], B[0], C[0]], [A[1], B[1], C[1]], color='peru')
ax[0][0].plot([A[0], B[0]], [A[1], B[1]], [A[0], C[0]], [A[1], C[1]], [C[0], B[0]], [C[1], B[1]], color='royalblue')
ax[0][0].text(A[0]+1, A[1]+1, f'A ({func(X, A[0], A[1])})', size=14)
ax[0][0].text(B[0], B[1]+5, f'B ({func(X, B[0], B[1])})', size=14)
ax[0][0].text(C[0]+1, C[1]+1, f'C ({func(X, C[0], C[1])})', size=14)
ax[0][0].set_title('Step 1', fontsize=15)

A2 = (3.5, 6)
R = (-16, -36)
L = (-9.5, -22)
ax[0][1].scatter([A2[0], B[0], C[0], A[0], R[0], L[0]], [A2[1], B[1], C[1], A[1], R[1], L[1]], color='peru')
ax[0][1].plot([A2[0], B[0]], [A2[1], B[1]], [A2[0], C[0]], [A2[1], C[1]], [C[0], B[0]], [C[1], B[1]], color='royalblue')
ax[0][1].plot([A[0], B[0]], [A[1], B[1]], [A[0], C[0]], [A[1], C[1]],
              [C[0], R[0]], [C[1], R[1]], [B[0], R[0]], [B[1], R[1]],
              [C[0], L[0]], [C[1], L[1]], [B[0], L[0]], [B[1], L[1]],
              linestyle='--', color='grey')
ax[0][1].text(A[0]+1, A[1]+1, f'A ({func(X, A[0], A[1])})', size=14)
ax[0][1].text(R[0]+2.5, R[1]-0.5, f'R ({func(X, R[0], R[1])})', size=14)
ax[0][1].text(L[0]+1, L[1]+2.5, f'L ({func(X, L[0], L[1])})', size=14)
ax[0][1].text(A2[0]+1, A2[1]+1, f'D ({func(X, A2[0], A2[1])})', size=14)
ax[0][1].text(B[0], B[1]+5, f'B ({func(X, B[0], B[1])})', size=14)
ax[0][1].text(C[0]+1, C[1]+1, f'C ({func(X, C[0], C[1])})', size=14)
ax[0][1].set_title('Step 2', fontsize=15)

C2 = (2.325, -6.75)
ax[1][0].scatter([A2[0], B[0], C2[0]], [A2[1], B[1], C2[1]], color='peru')
ax[1][0].plot([A2[0], B[0]], [A2[1], B[1]], [A2[0], C2[0]], [A2[1], C2[1]], [C2[0], B[0]], [C2[1], B[1]], color='royalblue')
ax[1][0].text(A2[0]+1, A2[1]+1, f'D ({func(X, A2[0], A2[1])})', size=14)
ax[1][0].text(B[0], B[1]+5, f'B ({func(X, B[0], B[1])})', size=14)
ax[1][0].text(C2[0]+1, C2[1]+1, f'E ({func(X, C2[0], C2[1])})', size=14)
ax[1][0].set_title('Step 3', fontsize=15)

C3 = (-2.4625, -1.625)
ax[1][1].scatter([A2[0], B[0], C3[0]], [A2[1], B[1], C3[1]], color='peru')
ax[1][1].plot([A2[0], B[0]], [A2[1], B[1]], [A2[0], C3[0]], [A2[1], C3[1]], [C3[0], B[0]], [C3[1], B[1]], color='royalblue')
ax[1][1].text(A2[0]+1, A2[1]+1, f'D ({func(X, A2[0], A2[1])})', size=14)
ax[1][1].text(B[0], B[1]+5, f'B ({func(X, B[0], B[1])})', size=14)
ax[1][1].text(C3[0]+1, C3[1]-1, f'K ({func(X, C3[0], C3[1])})', size=14)
ax[1][1].set_title('Step 4', fontsize=15)
plt.savefig('../../assets/images/optimization/nedler_mead_example.png', bbox_inches='tight');
```
