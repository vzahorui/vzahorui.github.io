```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm, uniform, laplace, norm

def plot_skewness():
    """
    Generates and plots a positively and negatively skewed distribution.
    """
    # Generate positively skewed data (right skew)
    # A log-normal distribution is a classic example of positive skew.
    pos_skew_data = np.random.lognormal(mean=0.5, sigma=1, size=1000)

    # Generate negatively skewed data (left skew)
    # A skewed normal distribution with a negative 'a' value.
    neg_skew_data = skewnorm.rvs(a=-10, size=1000)

    # Create a figure with two subplots side-by-side
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle('Visualizing Skewness', fontsize=16)

    # Plot the positively skewed distribution
    axes[0].hist(pos_skew_data, bins=50, density=True, alpha=0.7, color='skyblue')
    axes[0].set_title('Positive Skew (Right Skew)')
    axes[0].set_xlabel('Value')
    axes[0].set_ylabel('Frequency')
    axes[0].axvline(x=np.mean(pos_skew_data), color='red', linestyle='--', label='Mean')
    axes[0].axvline(x=np.median(pos_skew_data), color='green', linestyle='-', label='Median')
    axes[0].legend()
    axes[0].grid(True, which='both', linestyle='--', linewidth=0.5)

    # Plot the negatively skewed distribution
    axes[1].hist(neg_skew_data, bins=50, density=True, alpha=0.7, color='coral')
    axes[1].set_title('Negative Skew (Left Skew)')
    axes[1].set_xlabel('Value')
    axes[1].set_ylabel('Frequency')
    axes[1].axvline(x=np.mean(neg_skew_data), color='red', linestyle='--', label='Mean')
    axes[1].axvline(x=np.median(neg_skew_data), color='green', linestyle='-', label='Median')
    axes[1].legend()
    axes[1].grid(True, which='both', linestyle='--', linewidth=0.5)

    plt.tight_layout()
    plt.savefig('../assets/images/distributions/skewness.png', bbox_inches='tight')
    plt.show()

def plot_kurtosis():
    """
    Generates and plots distributions with different kurtosis levels.
    """
    # Mesokurtic (Normal) distribution - Kurtosis â‰ˆ 3
    meso_data = np.random.normal(loc=0, scale=1, size=5000)

    # Leptokurtic (Heavy tails) distribution - Kurtosis > 3
    # A Laplace distribution is an example of a leptokurtic distribution.
    lepto_data = laplace.rvs(loc=0, scale=0.5, size=5000)

    # Platykurtic (Light tails) distribution - Kurtosis < 3
    # A uniform distribution is an example of a platykurtic distribution.
    platy_data = uniform.rvs(loc=-2.5, scale=5, size=5000)

    # Create a figure with three subplots
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    fig.suptitle('Visualizing Kurtosis', fontsize=16)

    # Plot Leptokurtic distribution
    axes[0].hist(lepto_data, bins=50, density=True, alpha=0.7, color='purple', label='Leptokurtic')
    axes[0].set_title('Leptokurtic (Heavy Tails)')
    axes[0].set_xlabel('Value')
    axes[0].set_ylabel('Frequency')
    axes[0].legend()
    axes[0].grid(True, which='both', linestyle='--', linewidth=0.5)

    # Plot Mesokurtic distribution
    axes[1].hist(meso_data, bins=50, density=True, alpha=0.7, color='green', label='Mesokurtic')
    axes[1].set_title('Mesokurtic (Normal)')
    axes[1].set_xlabel('Value')
    axes[1].set_ylabel('Frequency')
    axes[1].legend()
    axes[1].grid(True, which='both', linestyle='--', linewidth=0.5)

    # Plot Platykurtic distribution
    axes[2].hist(platy_data, bins=50, density=True, alpha=0.7, color='orange', label='Platykurtic')
    axes[2].set_title('Platykurtic (Light Tails)')
    axes[2].set_xlabel('Value')
    axes[2].set_ylabel('Frequency')
    axes[2].legend()
    axes[2].grid(True, which='both', linestyle='--', linewidth=0.5)

    plt.tight_layout()
    plt.savefig('../assets/images/distributions/kurtosis.png', bbox_inches='tight')
    plt.show()
```

```python
plot_skewness()
```

```python
plot_kurtosis()
```
