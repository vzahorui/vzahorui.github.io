```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
def non_lin_func1(x):
    return 5/(4*x + 7)

def non_lin_func2(x):
    return np.sin(x/4)

x = np.linspace(0.001, 32, 300)

with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(1, 2, figsize=(18, 7))
ax[0].plot(x, non_lin_func1(x))
ax[0].text(9, 0.15, r'$f(x)=\frac{5}{4x+7}$', size=17)
ax[1].plot(x, non_lin_func2(x))
ax[1].text(11.5, 0.45, r'$\phi(x)=\sin(\frac{x}{4})$', size=17)
for ax in ax:
    ax.set_xlim(0, 30)
    ax.set_xlabel('x', fontsize=16)
    ax.set_ylabel('$y$', fontsize=16)

plt.savefig('../../assets/images/regression/non_linear_regression_examples.png', bbox_inches='tight');
```

# Line search material

```python
def func_quadr(x):
    return (x-25)**2+20

x = np.linspace(1, 45, 500)

with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(figsize=(12, 7))
    
ax.plot(x, func_quadr(x), color='royalblue', label=r'$f(x_k + \eta_k d_k)$')
ax.plot(x, -8*x + 525, linestyle='--', color='gray', label=r'$f(x_k) - c_1 \eta_k\nabla f(x_k)d_k$')
plt.legend(fontsize=15)

ax.set_xlabel(r'$\eta$', fontsize=16)
ax.set_ylabel(r'$f(x)$', fontsize=16)

ax.set_xlim(2.9, 45)

plt.savefig('../../assets/images/optimization/Armijo_rule.png', bbox_inches='tight');
```
