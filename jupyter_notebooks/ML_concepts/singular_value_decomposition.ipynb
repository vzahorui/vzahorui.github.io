```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(figsize=(10, 6))
    
ax.set_xlim(0, 10)
ax.set_ylim(0, 6)

ax.annotate(s='', xy=(0, 5), xytext=(0, 0), arrowprops=dict(width=2, color='indianred'))
ax.annotate(s='', xy=(8, 0), xytext=(0, 0), arrowprops=dict(width=2, color='blue'))
ax.annotate(s='', xy=(8, 5), xytext=(0, 0), arrowprops=dict(width=2, color='dimgrey'))

ax.plot([0, 8], [5, 5], color='grey', linestyle='--', dashes=(10, 8))
ax.plot([8, 8], [0, 5], color='grey', linestyle='--', dashes=(10, 8))

ax.annotate(s='$\overrightarrow{a}$', xy=(0, 0), xytext=(4, 3), fontsize=18)
ax.annotate(s='$\overrightarrow{v_2}$', xy=(0, 0), xytext=(-0.7, 2.5), fontsize=18)
ax.annotate(s='$\overrightarrow{v_1}$', xy=(0, 0), xytext=(4, -0.7), fontsize=18)
ax.annotate(s='$s_2$', xy=(0, 0), xytext=(8.2, 2.5), fontsize=18)
ax.annotate(s='$s_1$', xy=(0, 0), xytext=(4, 5.2), fontsize=18)

plt.savefig('../../assets/images/concepts/simple_vector_decomposition.png', bbox_inches='tight');
```
