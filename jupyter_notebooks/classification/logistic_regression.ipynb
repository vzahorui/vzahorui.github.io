```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
def logistic_function(t):
    f_t = 1/(1+np.exp(-t))
    return f_t
```

```python
with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(figsize=(7, 4))
t = np.linspace(-6, 6, 100)    
ax.plot(t, logistic_function(t), zorder=5)
ax.axvline(0, linestyle='--', color='silver', zorder=0)
plt.tight_layout()
ax.set_xlabel(r'$t$')
ax.set_ylabel(r'$f(t)$')
plt.savefig('../../assets/images/regression/logistic_function.png', bbox_inches='tight');
```
