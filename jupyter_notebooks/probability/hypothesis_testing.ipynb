```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, ttest_1samp

%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize=(12,6))
x_start, x_end = 1, 9
x = np.arange(x_start, x_end, 0.001)
mu = 5
sigma = 1
y = norm.pdf(x, mu, sigma)
ax.plot(x, y, color='royalblue', alpha=0.9)
x1 = np.arange(7.2, x_end, 0.001)
y1 = norm.pdf(x1, mu, sigma)
ax.fill_between(x1, y1, 0, alpha=0.2, color='b')
x2 = np.arange(7.4, x_end, 0.001)
y2 = norm.pdf(x2, mu, sigma)
ax.fill_between(x2, y2, 0, alpha=0.3, color='r')
ax.axvline(mu, color='darkgrey', linestyle='--')
ax.annotate('$\mu$', xy=(0.51, -0.04), ha='center', va='center', xycoords='axes fraction', fontsize=15)
ax.annotate(text='p-value', xy=(7.7, 0.01), xytext=(7.9, 0.08), 
            arrowprops=dict(color='dimgrey', arrowstyle='->'),
            fontsize=13)
ax.annotate(text=r'$\alpha$', xy=(7.3, 0.026), xytext=(7.7, 0.09), 
            arrowprops=dict(color='dimgrey', arrowstyle='->'),
            fontsize=13) 
ax.get_yaxis().set_visible(False)
ax.get_xaxis().set_visible(False)
ax.set_xlim(x_start, x_end)
plt.title('Significance level and p-value when regecting the null hypothesis', fontsize=15)
plt.tight_layout()
plt.savefig('../../assets/images/probability/significance_level_demo.png', bbox_inches='tight');
```

```python
fig, ax = plt.subplots(figsize=(12,6))
x_start, x_end = 1, 14
x = np.arange(x_start, x_end, 0.001)
mu = 5
mu_2 = 9
sigma = 1
y = norm.pdf(x, mu, sigma)
y_2 = norm.pdf(x, mu_2, sigma)
ax.plot(x, y, color='royalblue', alpha=0.9)
ax.plot(x, y_2, color='firebrick', alpha=0.9)
ax.fill_between(x, y, 0, alpha=0.2, color='b')
ax.fill_between(x, y_2, 0, alpha=0.2, color='r')
ax.axvline(mu, color='darkgrey', linestyle='--')
ax.annotate('$\mu_0$', xy=(0.325, -0.04), ha='center', va='center', xycoords='axes fraction', fontsize=15)
ax.axvline(mu_2, color='darkgrey', linestyle='--')
ax.annotate('$\mu_1$', xy=(0.635, -0.04), ha='center', va='center', xycoords='axes fraction', fontsize=15)
# significance level for H0
z_0 = 7.2
x_s = np.arange(z_0, x_end, 0.001)
y_s = norm.pdf(x_s, mu, sigma)
ax.fill_between(x_s, y_s, 0, alpha=0.4, color='b')
ax.annotate(text=r'$\alpha$', xy=(7.35, 0.015), xytext=(7.9, 0.06), 
            arrowprops=dict(color='black', arrowstyle='->'),
            fontsize=13)
ax.axvline(z_0, color='darkgrey', linestyle='--')
ax.annotate('$x_0$', xy=(0.50, -0.04), ha='center', va='center', xycoords='axes fraction', fontsize=15)
# area for type II error
x_t2 = np.arange(x_start, z_0, 0.001)
y_t2 = norm.pdf(x_t2, mu_2, sigma)
ax.fill_between(x_t2, y_t2, 0, alpha=0.4, color='r')
ax.annotate(text='Area for type II eror', xy=(6.9, 0.028), xytext=(7.8, 0.15), 
            arrowprops=dict(color='black', arrowstyle='->'),
            fontsize=13)
ax.get_yaxis().set_visible(False)
ax.get_xaxis().set_visible(False)
ax.set_xlim(x_start, x_end)
plt.tight_layout()
plt.savefig('../../assets/images/probability/distributions_overlap.png', bbox_inches='tight');
```

# Confidence interval

```python
fig, ax = plt.subplots(figsize=(12,6))
x_start, x_end = -4, 4
x = np.arange(x_start, x_end, 0.001)
mu = 0
sigma = 1
y = norm.pdf(x, mu, sigma)
x_1 = np.arange(x_start, -1.96, 0.001)
y_1 = norm.pdf(x_1, mu, sigma)
x_2 = np.arange(1.96, x_end, 0.001)
y_2 = norm.pdf(x_2, mu, sigma)
ax.plot(x, y, color='royalblue', alpha=0.9)
ax.fill_between(x_1, y_1, 0, alpha=0.2, color='r')
ax.fill_between(x_2, y_2, 0, alpha=0.2, color='r')
ax.axvline(mu, color='darkgrey', linestyle='--')
ax.annotate(r'$\bar x$', xy=(0.51, -0.04), ha='center', va='center', xycoords='axes fraction', fontsize=15)
ax.annotate(text=r'$\frac{\alpha}{2}$', xy=(2.2, 0.02), xytext=(2.5, 0.06), 
            arrowprops=dict(color='black', arrowstyle='->'),
            fontsize=15)
ax.annotate(text=r'$\frac{\alpha}{2}$', xy=(-2.2, 0.02), xytext=(-2.5, 0.06), 
            arrowprops=dict(color='black', arrowstyle='->'),
            fontsize=15)
ax.annotate(text='', xy=(-1.96, 0.0), xytext=(0, 0), arrowprops=dict(color='black', arrowstyle='<->'))
ax.annotate(text='', xy=(1.96, 0.0), xytext=(0, 0), arrowprops=dict(color='black', arrowstyle='<->'))
ax.text(0.8, 0.01, r'$Z \cdot SE$', fontsize=13)
ax.text(-1.1, 0.01, r'$Z \cdot SE$', fontsize=13)
ax.get_yaxis().set_visible(False)
ax.get_xaxis().set_visible(False)
ax.set_xlim(x_start, x_end)
plt.tight_layout()
plt.savefig('../../assets/images/probability/confidence_interval.png', bbox_inches='tight');
```

# Vizualizations for False discovery method

```python
def simulate_tests(mu, sample_mean, sample_size = 300, tests = 200):
    '''Run multiple tests and return p-values for each test'''
    p_values = []
    counter = 0
    while counter < tests:
        sample = norm.rvs(loc=sample_mean, size=sample_size, random_state=counter+23)
        z_value, p_value = ttest_1samp(sample, mu)
        p_values.append(p_value)
        counter += 1
    return p_values
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14, 6))
p_values_same_mean = simulate_tests(10, 10, tests=9000)
sns.histplot(p_values_same_mean, bins=10, ax=ax[0])
p_values_dif_mean = simulate_tests(10.1, 10, tests=1000)
sns.histplot(p_values_dif_mean, bins=10, ax=ax[1])
ax[0].set_title('$H_0$ is true', fontsize=15)
ax[1].set_title('$H_0$ is false', fontsize=15)
for ax in ax:
    ax.set(ylabel=None)
plt.tight_layout()
plt.savefig('../../assets/images/probability/p_values_distribution.png', bbox_inches='tight');
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
p_values_combined = p_values_same_mean + p_values_dif_mean
sns.histplot(p_values_combined, bins=10)
ax.set_xlabel(r'$p$-values', fontsize=13)
ax.set_ylabel('Frequency', fontsize=13)
plt.tight_layout()
plt.savefig('../../assets/images/probability/combined_p_values_distribution.png', bbox_inches='tight');
```

```python
with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(1, 2, figsize=(14, 6), sharey=True)
p_values_same_mean_sorted = p_values_same_mean.copy()
p_values_same_mean_sorted.sort()
p_values_combined_sorted = p_values_combined.copy()
p_values_combined_sorted.sort()
ax[0].plot(range(len(p_values_same_mean)), p_values_same_mean_sorted, label='Sorted $p$-values', 
           linewidth=4, color='cornflowerblue')
ax[1].plot(range(len(p_values_combined_sorted)), p_values_combined_sorted, label='Sorted $p$-values', 
           linewidth=4, color='cornflowerblue')

x_slope = np.array(range(len(p_values_same_mean)))
y_slope = x_slope/len(p_values_same_mean)
ax[0].plot(x_slope, y_slope, label='Slope approximation', color='maroon', linewidth=2, linestyle='--')
x_slope2 = np.array(range(len(p_values_combined_sorted)))
y_slope2 = x_slope2/len(p_values_combined_sorted)
ax[1].plot(x_slope2, y_slope2, label='Slope approximation', color='maroon', linewidth=2, linestyle='--')

ax[0].legend(fontsize=12)
ax[0].set_ylabel('p-value', fontsize=13)
ax[0].set_title('All hypotheses are true', fontsize=15)
ax[1].set_title('Combined true and false hypotheses', fontsize=15)
for a in ax:
    a.set_xlabel('$k$', fontsize=13)  
plt.tight_layout()
plt.savefig('../../assets/images/probability/sorted_p_values.png', bbox_inches='tight');
```

```python
with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(figsize=(10, 6))

y_slope2_adj = y_slope2*0.1
ax.plot(range(len(p_values_combined_sorted)), p_values_combined_sorted, linewidth=4, color='cornflowerblue')
ax.plot(x_slope2, y_slope2_adj, color='maroon', linewidth=2, linestyle='--')
ax.set_ylabel('p-value', fontsize=13)
ax.set_xlabel('$k$', fontsize=13)  
plt.tight_layout()
plt.savefig('../../assets/images/probability/false_dsicovery_rate_viz.png', bbox_inches='tight');
```
