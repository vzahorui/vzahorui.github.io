```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, beta

%matplotlib inline
```

```python
n = 287
k = 123

x = np.linspace(0, 1.01, 1000)
binom_dist = binom.pmf(k=k, n=n, p=x)
prior_dist = beta.pdf(x=x, a=60, b=40)
posterior_dist = prior_dist*binom_dist

with sns.axes_style('whitegrid'):
    fig, ax = plt.subplots(nrows=3, figsize=(12,9))

ax[0].plot(x, prior_dist)
ax[1].plot(x, binom_dist, label='Likelihood function')
ax[2].plot(x, posterior_dist, label='Posterior function')

ax[0].set_title('Prior hypothesis distribution', fontsize=14)
ax[1].set_title('Likelihood function distribution', fontsize=14)
ax[2].set_title('Posterior hypothesis distribution', fontsize=14)

ax[0].text(0.21, 4.8, s=r'$\alpha=60$, $\beta=40$', fontsize=14)



for ax in ax:
    ax.set_xlim(0, 1)
        
plt.tight_layout()
plt.savefig('../../assets/images/probability/posterior_distribution_via_map.png', bbox_inches='tight');
```
