```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def convex_function(x):
    """
    A simple convex function for our example: f(x) = x^2.
    """
    return x**2

def main():
    """
    Main function to illustrate Jensen's Inequality.
    """
    # 1. Generate a set of random numbers to represent our variable X
    x_values = np.array([-2, 3])
    
    # 2. Calculate the Expected Value (mean) of X
    expected_x = np.mean(x_values)
    
    # 3. Calculate the function of the expected value: f(E[X])
    f_of_expected_x = convex_function(expected_x)
    
    # 4. Calculate the Expected Value of the function: E[f(X)]
    f_x_values = convex_function(x_values)
    expected_f_of_x = np.mean(f_x_values)

    # 6. Plot the results to visually demonstrate the inequality
    with sns.axes_style("whitegrid"):
        fig, ax = plt.subplots(figsize=(10, 6))

    # Plot the convex function curve
    x_plot = np.linspace(-5, 5, 400)
    y_plot = convex_function(x_plot)
    ax.plot(x_plot, y_plot, label='$f(x) = x^2$')

    # Plot the original data points on the curve
    ax.scatter(x_values, f_x_values)
    
    # Connect the data points with a line to show their average
    ax.plot(x_values, f_x_values, label='')

    # plot mean values of the function
    ax.scatter([expected_x], [expected_f_of_x], label='$E[f(X)]$')
    ax.scatter([expected_x], [f_of_expected_x], label='$f(E[X])$')

    ax.plot([x_values[0], x_values[0]], [0, f_x_values[0]], 'k--', alpha=0.5)
    ax.plot([x_values[1], x_values[1]], [0, f_x_values[1]], 'k--', alpha=0.5)
    ax.text(x_values[0], y=0.5, s='$X_1$', fontsize=12, ha='right')
    ax.text(x_values[1], y=0.5, s='$X_2$', fontsize=12, ha='left')
    ax.hlines(y=0, xmin=x_plot[0]*1.1, xmax=x_plot[-1]*1.1, color='darkgrey')
    
    
    ax.set_title("Visualizing Jensen's Inequality", fontsize=16)
    ax.set_xlabel("$x$")
    ax.set_ylabel("$f(x)$")
    ax.legend()
    ax.grid(True, linestyle='--')
    plt.savefig('../../assets/images/probability/jensen_inequality_demo.png', bbox_inches='tight');
    plt.show()

main()
```
